```python
import numpy as np
import matplotlib.pyplot as plt
```

# Parameters

```python
# parameters
S0 = 305
R = 10 ** 6 #Initial number of stocks
T = 1; 
k = 0.0004;
b = 0.0004
M = 2000;    #Number of trading times
alpha = 0.1; 
phi = 0.01;
sigma = 0.4;
```

# Rest Parameters

```python
gamma = np.sqrt(phi/k);
xi = (alpha - 0.5 * b + np.sqrt(k * phi))/(alpha - 0.5 * b - np.sqrt(k * phi));
```

```python
tt = np.linspace(0, 1, M + 1)
```

# Optimal trading speed

```python
v = np.zeros((M + 1, 1))
q = np.zeros((M + 1, 1))
v[0] = 0
q[0] = R  
for t in range(1, M + 1): 
    v[t] = gamma * (xi * np.exp(gamma * (T-tt[t-1])) + np.exp(- gamma * (T-tt[t-1]))) / (xi * np.exp(gamma * (T-tt[t-1])) - np.exp(- gamma * (T-tt[t-1]))) * q[t-1] * T/M
    q[t] = (xi * np.exp(gamma * (T-tt[t-1])) - np.exp(- gamma * (T-tt[t-1]))) / (xi * np.exp(gamma * T) - np.exp(- gamma * T)) * R
```

```python
plt.figure(figsize=(10, 6))
plt.plot(tt[1:], v[1:], 'b', lw=1.5)
plt.xlabel('Time')
plt.ylabel('Number of Shares')
plt.title('Trading Speed')
```

# Inventory Process

```python
plt.figure(figsize=(10, 6))
plt.plot(tt, q, 'b', lw=1.5)
plt.xlabel('Time')
plt.ylabel('Number of Shares')
plt.title('Inventory')
```

# Simulation of stock price

```python
S = np.zeros((M + 1, 1))
S[0] = S0 
rn = np.random.standard_normal(S.shape) 
for t in range(1, M + 1): 
    S[t] = S[t-1] - b * v[t] * T/M + sigma * np.sqrt(T/M) * rn[t]
```

```python
plt.figure(figsize=(10, 6))
plt.plot(tt, S, 'b', lw=1.5)
plt.xlabel('Time')
plt.ylabel('$')
plt.title('Stock Price')
```

# Revenue Generation process

```python
R = np.zeros((M + 1, 1))
R[0] = 0
for t in range(1, M + 1): 
    R[t] = R[t-1] +  (S[t] - k * v[t]) * v[t]
```

```python
plt.figure(figsize=(10, 6))
plt.plot(tt, R, 'b', lw=1.5)
plt.xlabel('Time')
plt.ylabel('$')
plt.title('Revenue')
```
